```python
import matplotlib
from matplotlib import pyplot as plt
import numpy as np
import GPy as gpy
from IPython.display import display
```

```python
# (python does not typically show plots until you ask it, so this inline call gives more 'natural' behavior)
%matplotlib inline  
%config InlineBackend.figure_format = 'svg'
matplotlib.rcParams['figure.figsize'] = (8,5)
```

```python
num_points = 200
x = np.random.uniform(-1.,1.,(num_points,1))
y = np.sin(x*2*np.pi) + 0.1*np.random.randn(num_points,1)
plt.plot(x,y,'kx',markersize=2,markeredgewidth=1)
```

```python
kernel = gpy.kern.RBF(input_dim=1, variance=1., lengthscale=1.)
```

```python
kernel2 = gpy.kern.RBF(input_dim=1, variance=1., lengthscale=10.)
```

```python
m = gpy.models.GPRegression(x,y,kernel)
display(m)
n=gpy.models.GPRegression(x,y,kernel2)
display(n)
```

```python
m.plot()
n.plot()
```

```python
m.optimize_restarts(num_restarts = 10)
n.optimize_restarts(num_restarts=10)
```

```python
display(m)
display(n)
```

```python
m.plot()
n.plot()
```
